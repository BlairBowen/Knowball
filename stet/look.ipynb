````python
import pandas as pd
from IPython.display import display, HTML, Markdown

def show_scrollable_df(df, height=400):
    display(HTML(df.to_html(notebook=True, max_rows=None, max_cols=None, border=0, escape=False, index=True, justify='center',
                            table_id='scroll_df')))
    display(HTML(f"""
    <style>
        #scroll_df {{
            display: block;
            overflow-y: scroll;
            height: {height}px;
            border: 1px solid #ddd;
            font-family: monospace;
        }}
    </style>
    """))
````

````python
nba_df = pd.read_csv("nba_players_with_obscurity.csv")
nfl_df = pd.read_csv("nfl_players_with_obscurity.csv")
epl_df = pd.read_csv("epl_players_with_obscurity.csv")
````

````python
nba_df.columns
````

````python
total_nba = nba_df[["Player", "Team", "obscurity_score", "obscurity_score_normalized", "obscurity_score_scaled"]][:].sort_values(by="obscurity_score")
````

````python
nfl_df.columns
````

````python
total_nfl = nfl_df[["Player", "Team", "Pos", "obscurity_score", "obscurity_score_normalized", "obscurity_score_scaled"]][:].sort_values(by="obscurity_score")
````

````python
epl_df.columns
````

````python
total_epl = epl_df[["Player", "Team", "obscurity_score", "obscurity_score_normalized", "obscurity_score_scaled"]][:].sort_values(by="obscurity_score")
````

````python
# display(Markdown(f"```\n{total_nba.to_string(index=True)}\n```"))
show_scrollable_df(total_nba)
show_scrollable_df(total_nfl.loc[total_nfl["Pos"] == "QB"])
# show_scrollable_df(total_nfl)
# show_scrollable_df(total_nfl.groupby("Pos").mean(numeric_only=True).reset_index().sort_values(by="obscurity_score"))
show_scrollable_df(total_epl)
````
